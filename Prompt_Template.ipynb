```python
from langchain_core.output_parsers import PydanticOutputParser
from pydantic import BaseModel, Field

class CarSpecification(BaseModel):
    brand: str = Field(description="Car brand name")
    model: str = Field(description="Car model name")
    fuel_type: str = Field(description="Fuel type (Petrol/Diesel/Electric/Hybrid)")
    transmission: str = Field(description="Transmission type (Manual/Automatic)")
    mileage_kmpl: float = Field(description="Mileage in km per liter")
    price_lakhs: float = Field(description="Price of the car in lakhs")
    description: str = Field(description="Detailed car specification overview")
    is_available: bool = Field(description="Is the car currently available for purchase")

# 2. Output Parser

parser = PydanticOutputParser(pydantic_object=CarSpecification)

# 3. Prompt Template

prompt = ChatPromptTemplate.from_template(
    """
    Generate a mock car specification for {query}.

    {format_instructions}
    """
)

# 4. Chain

chain = prompt | llm | parser


# 5. Invoke

try:
    result = chain.invoke({
        "query": "Hyundai Creta for family use in city driving",
        "format_instructions": parser.get_format_instructions()
    })

    print("Brand:", result.brand)
    print("Model:", result.model)
    print("Fuel Type:", result.fuel_type)
    print("Transmission:", result.transmission)
    print("Mileage (kmpl):", result.mileage_kmpl)
    print("Price (Lakhs):", result.price_lakhs)
    print("Description:", result.description)
    print("Available:", result.is_available)

    if result.price_lakhs > 20:
        print("miru ante vunnollu bro!")

except Exception as e:
    print("Parsing Error:", e)
```

```python
from langchain_core.output_parsers import PydanticOutputParser
from pydantic import BaseModel, Field
from typing import List


class FoodTweetReview(BaseModel):
    content: str = Field(description="Funny viral tweet about food, max 300 characters")
    hashtags: List[str] = Field(description="Relevant and trending food hashtags")
    sentiment: str = Field(description="Overall sentiment (Positive/Negative/Mixed)")


parser = PydanticOutputParser(pydantic_object=FoodTweetReview)

prompt = ChatPromptTemplate.from_template(
    """
    Write a funny and viral Twitter review about eating {food}
    at {place}. Use humor and exaggeration.

    {format_instructions}
    """
)

chain = prompt | llm | parser

review = chain.invoke({
        "food": "pani puri",
        "place": "a busy street stall",
        "format_instructions": parser.get_format_instructions()
    })

print(review)
```
